## 1.3.Tipo Raw

Esta celda se utiliza cuando se busca que Jupyter no interprete el contenido de la misma y lo muestre tal como se escribió no tienen distinción de entrada-salida, son celdas sin formato. Esto significa que las celdas sin procesar no se pueden convertir en nada más que lo que ya son. Si se hace clic en el botón Ejecutar, la celda permanecerá tal cual está y el Notebook automáticamente selecciona justo la celda debajo de ella.

# 2.Estilo de Fuente

Se puede resaltar distintivamente un texto.

## 2.1.Fuente Itálica

Sirve para poner en cursiva el texto, se pone * * entre la frase a resaltar.

Ejemplo: *Escuela Nacional de Ciencias de la Tierra*

## 2.2.Fuente Negrita

Sirve para engrosar el texto, se pone ** ** entre la frase a resaltar.

Ejemplo: **Escuela Nacional de Ciencias de la Tierra**

## 2.3.Fuente Negrita e Itálica

Sirve para poner en cursiva y engrosar el texto, se pone *** *** entre la frase a resaltar.

Ejemplo: ***Escuela Nacional de Ciencias de la Tierra***
